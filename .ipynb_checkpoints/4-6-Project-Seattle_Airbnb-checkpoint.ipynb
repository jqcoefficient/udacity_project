# I. Correlation between "review_scores_rating" and other ratings

To understand the correlation between overall listing rating and the other categories of rating, we can use a scatter plot to see if there's a general trend before diving into deeper analysis. So, first, let's plot "review_scores_rating" vs each of the other ratings to get a high level understanding. 

We start by importing the necessary Python packages. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr
```

## Gather Data

Next, we gather the necessary data for analysis. The original .csv file has a lot of data. We don't need all of them. In addition, if we run our script over the original .csv file, it will take unnecessary computation time. 

We first load the .csv file into a Pandas dataframe. Then, we create a list of the coloumns we want to look at. 

```python
df = pd.read_csv("listings.csv")
rating_col = "review_scores_rating"
other_col = ["review_scores_accuracy",
    "review_scores_cleanliness",
    "review_scores_checkin",
    "review_scores_communication",
    "review_scores_location",
    "review_scores_value"]
```

## Clean Data

There are some NaN and/or blank values in the aforementioned columns. In order to perform the scipy.stats methods we need to get rid of the NaN. We can use the imputation method of dropping the NaN rows because we have plenty of data. 

After data cleaning, we store the cleaned dataset in a new dataframe and work with this smaller dataframe.

```python
all_rating_col = other_col + [rating_col]
df_clean_ratings = df.dropna(subset=all_rating_col)
```

## Data Exploration

We can now do data exploration and look at the data we filtered. We loop each column and plot it in a subplot. 

```python
fig, axes = plt.subplots(2, 3, figsize=(15,10))
axes = axes.flatten()

for i, col in enumerate(other_col):
    ax = axes[i]
    ax.scatter(df_clean_ratings[col], df_clean_ratings[rating_col])
    ax.set_xlabel(col)
    ax.set_ylabel(rating_col)

plt.tight_layout()
plt.show()
```

## Data Analysis and Modeling

Looks like all six plots show a general positive correlation between the overall rating and the subcategories. Now, let's dive deeper into the data analysis by finding the correlation coefficient for each dataset and plotting the linear fit line. We can include the correlation value in the title of each plot. 

Also, since we foresee calculating the correlation coefficient and drawing the linear fit line for other analyses in this project, let's create a function for that. Let's call it "corr_and_line."

```python
def corr_and_line(x, y):
    """This function takes in two columns of data and returns the correlation coefficient, 
    the p_value, and two arries of x and y to plot the linear fit line."""
    
    corr, p_value = pearsonr(x, y)
    slope, intercept = np.polyfit(x, y, 1)
    line_x = np.linspace(x.min(), x.max(), 100)
    line_y = slope * line_x + intercept
    return corr, p_value, line_x, line_y

for col in other_col:
    x = df_clean_ratings[col]
    y = df_clean_ratings[rating_col]
    
    corr, p_value, line_x, line_y = corr_and_line(x, y)
    
    plt.figure(figsize=(7,5))
    plt.scatter(x, y)
    plt.plot(line_x, line_y)
        
    plt.xlabel(col)
    plt.ylabel(rating_col)
    plt.title(f"{rating_col} vs {col}\nCorrelation: {corr:.3f}")
    plt.show()
```

## Results and Discussion 

Here are the results. 
* Overall rating vs listing accuracy has a correlation coefficient of 0.622. This is moderately high. This means guests care about the listing accuracy.
* Overall rating vs cleanliness has a correlation coefficient of 0.642. This is moderately high. This means guests care about the cleanliness.
* Overall rating vs check-in has a correlation coefficient of 0.522. This is moderate. This means guests may or may not care about the check-in experience.
* Overall rating vs communication has a correlation coefficient of 0.544. This is moderate. This means guests may or may not care about the communication.
* Overall rating vs location has a correlation coefficient of 0.369. This is low. This means guests most likely will not see location as an important factor in providing the overall rating.
* Overall rating vs value has a correlation coefficient of 0.696. This is the highest correlation here. This means guests see the value they get out of the Airbnb as important.

# II. Correlation between "host_response_rate" and "review_scores_communication"

The next question we want to know the answer to is whether the host's response rate affect the communication rating. If it does, the hosts can improve the overall rating by simply being more tentative to messages from guests. If it does not, the hosts can divert the time and energy elsewhere to improve the overall experience of the rental.

## Gather and Clean Data

We start off by clean the data by getting rid of the NaN rows. We can do that because there is overwhelming amount of data that have values in them. Similar to Section I, we create a smaller dataframe with just the necessary columns. As we explore the data, we realized that the host response rate is captured as a string value such as "100%." We need to do some further data cleaning to convert it to a float data type in order to perform statistics on the data. 

```python
comms = ["host_response_rate", "review_scores_communication"]
df_clean_comms = df.dropna(subset=comms)
```

So, first we ensure that both columns of data are in the float data type. Then, we find the correlation and plot the relationship.

```python
x = df_clean_comms[comms[0]]
y = df_clean_comms[comms[1]]

# "host_response_rate" has a data type of 'object' since the values are in formats like this: 100%. 
# It is needed to convert it to a data type of 'float' just like the data type in "review_scores_communication."

x = x.str.replace('%', '', regex=False)
x = x.astype(float)
```

## Data Analysis and Modeling

```python
corr, p_value, line_x, line_y = corr_and_line(x, y)

plt.figure(figsize=(7,5))
plt.scatter(x, y)
plt.plot(line_x, line_y)
    
plt.xlabel(comms[0])
plt.ylabel(comms[1])
plt.title(f"{comms[1]} vs {comms[0]}\nCorrelation: {corr:.3f}")
plt.show()
```

This is surprising! There's a very low correlation between te host's response rate and the communication score. This suggests that host's response rate has a low influence on the communciation score.

# III. Correlation between "square_feet" and price

What the property's square footage effect on the listing price? Does square footage drive price? Let's find out.

## Gather and Clean Data

Like bofore, we first clean the data of any NaNs and inconsistent data types. Then we create a working dataframe containing only the necessary data.

```python
size_and_price = ["square_feet", "price"]
df_clean_size_price = df.dropna(subset=size_and_price)

x = df_clean_size_price[size_and_price[0]]
y = df_clean_size_price[size_and_price[1]]

# "price" has a data type of 'object' since the values are in formats like this: $300.00. 
# It is needed to convert it to a data type of 'float' just like the data type in "square_feet"

y = y.str.replace('$', '', regex=False)
y = y.astype(float)
```

## Data Analysis and Modeling

Next, we simply find the correlation and plot the relationship. 

```python
corr, p_value, line_x, line_y = corr_and_line(x, y)

plt.figure(figsize=(7,5))
plt.scatter(x, y)
plt.plot(line_x, line_y)
    
plt.xlabel(size_and_price[0])
plt.ylabel(size_and_price[1])
plt.title(f"{size_and_price[1]} vs {size_and_price[0]}\nCorrelation: {corr:.3f}")
plt.show()
```

Looks like there's a moderate correlation between the size of the house and the listing price. This is also surprising because one would expect a high correlation between size of the house and the price. Maybe there are other factors, such as seasonality and location, that could affect this correlation. This could be explored in the next project. 

# Conclusion

Theres a moderate to high correlation between value of the rental, cleanliness, and accuracy of the listing to the overall rating. Theres a moderate correlation between the check-in process and communication to the overall rating. Theres a weak correlation between the location of the listing to the overall rating. Also, there an extremely weak correlation between the hosts response rate to the communication score. As expected, theres is positive moderate correlation between the square footage of the listing to the price.
